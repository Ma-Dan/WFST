```python
import numpy as np
import openfst_python as fst
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
def ReadFstStringFile(file_name):
    compiler = fst.Compiler()

    f = open(file_name, 'r')
    lines = f.readlines()
    for line in lines:
        compiler.write(line)
    f.close()

    return compiler.compile()        
```

```python
a = ReadFstStringFile('A.fst')
a
```

```python
b = ReadFstStringFile('B.fst')
b
```

```python
c = fst.compose(a, b)
c
```
