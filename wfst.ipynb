```python
import numpy as np
import openfst_python as fst
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
a = fst.Fst()
s0 = a.add_state()
s1 = a.add_state()
s2 = a.add_state()
s3 = a.add_state()
a.add_arc(s0, fst.Arc(1, 2, fst.Weight(a.weight_type(), 0.1), s1))
a.add_arc(s0, fst.Arc(2, 1, fst.Weight(a.weight_type(), 0.2), s2))
a.add_arc(s1, fst.Arc(3, 1, fst.Weight(a.weight_type(), 0.3), s1))
a.add_arc(s1, fst.Arc(1, 1, fst.Weight(a.weight_type(), 0.4), s3))
a.add_arc(s2, fst.Arc(2, 2, fst.Weight(a.weight_type(), 0.5), s3))
a.set_start(s0)
a.set_final(s3, fst.Weight(a.weight_type(), 0.6))
a
```

```python
b = fst.Fst()
s0 = b.add_state()
s1 = b.add_state()
s2 = b.add_state()
b.add_arc(s0, fst.Arc(2, 3, fst.Weight(b.weight_type(), 0.3), s1))
b.add_arc(s1, fst.Arc(1, 2, fst.Weight(b.weight_type(), 0.4), s2))
b.add_arc(s2, fst.Arc(1, 2, fst.Weight(b.weight_type(), 0.6), s2))
b.set_start(s0)
b.set_final(s2, fst.Weight(a.weight_type(), 0.7))
b
```

```python
c = fst.compose(a, b)
c
```
